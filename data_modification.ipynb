```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_desmd=pd.read_csv("desmedt2007_genomicMatrix _coded.csv")
file_mapped=pd.read_csv("geneNames.csv")
file_desmd.shape
```

```python
file_mapped=file_mapped.to_numpy()
file_mapped.shape
```

```python

probe_ids=[]
gene_ids=[]

for a in range(file_mapped.shape[0]):
    for b in range(1):
        probe_ids.append(file_mapped[a][b])
        gene_ids.append(file_mapped[a][b+1])


print(probe_ids)
print(gene_ids)
print(len(probe_ids), len(gene_ids))
```

```python
probe_ids[17],probe_ids[18],probe_ids[19]
```

```python
filtered_prob_ids=[]
filtered_gene_ids=[]
for i in range(len(probe_ids)):
    if probe_ids.count(probe_ids[i]) == 1:
        filtered_prob_ids.append(probe_ids[i])
        filtered_gene_ids.append(gene_ids[i])
print(filtered_prob_ids)
print(filtered_gene_ids)
print(len(filtered_prob_ids), len(filtered_gene_ids))
```

```python
file_desmd=file_desmd.to_numpy()
file_desmd[0]
```

```python
condition=np.isin(file_desmd[:,0],filtered_prob_ids)
filtered_desmd=file_desmd[condition]
```

```python
filtered_probs=[]
filtered_genes=[]
X=filtered_desmd[:,0]

import math

def is_not_nan(value):
    if not isinstance(value, float):
        return True

    return not math.isnan(value)

for i in range(len(filtered_prob_ids)):
    if is_not_nan(filtered_gene_ids[i]) and filtered_prob_ids[i] in X:
        filtered_probs.append(filtered_prob_ids[i])
        filtered_genes.append(filtered_gene_ids[i])

print(len(filtered_probs),len(filtered_genes))
```

```python
condition=np.isin(filtered_desmd[:,0],filtered_probs)
filtered_desmd=filtered_desmd[condition]
len(filtered_desmd)
```

```python
'''
filtered_probs = [a, b, c]
filtered_genes = [x, x, z]
file_desmd = [
    [a, 1, 2, 3],
    [b, 1, 2, 3],
    [c, 1, 2, 3],
    [d, 1, 2, 3],
]
# output
filtered_desmd = [
    [a, 1, 2, 3],
    [b, 1, 2, 3],
    [c, 1, 2, 3],
]
# average
file_desmd_average = [
    [x, 1.5, 2.5, 3.5], # average 
    [z, 1, 2, 3],
]
'''
file_desmd.shape, filtered_desmd.shape, len(filtered_probs), len(filtered_genes)
```

```python
'''
- x = [0 ,1] indexes => [a,b]
- find a and b from filtered_desmd, find average
- add new row with [x, average_value1, average_value2, ...]
'''
index_map = {}

for i in range(len(filtered_genes)):
  val = filtered_genes[i]
  if val not in index_map:
    index_map[val] = []
    index_map[val].append(i)
  else:
    index_map[val].append(i)

# print(index_map)
filtered_desmd_average = []

for _, (key, value) in enumerate(index_map.items()):
  # print(f"Key: {key}, Value: {value}")
  selected_indexes = []
  for index in value:
    f_prob = filtered_probs[index]
    selected_indexes.append(np.where(filtered_desmd[:, 0] == f_prob)[0][0])

  # print(selected_indexes)
  data_modified = filtered_desmd.copy()
  data_modified[selected_indexes, 0] = key

  average_rows = np.mean(data_modified[selected_indexes, 1:], axis=0)
  average_rows = average_rows.tolist()
  average_rows.insert(0, key)
  # print("row average: ",average_rows)
  # print(data_modified)
  filtered_desmd = data_modified
  # print(filtered_desmd)
  filtered_desmd_average.append(average_rows)


print('------- Original -------')
print(filtered_desmd)
# print(filtered_desmd_average)
filtered_desmd_average = np.array(filtered_desmd_average)
print('------- New -------')
print(filtered_desmd_average)
```
